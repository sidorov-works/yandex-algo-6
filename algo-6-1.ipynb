### A. Плот

Посередине озера плавает плот, имеющий форму прямоугольника. Стороны плота направлены вдоль параллелей и меридианов. 

Введём систему координат, в которой ось $OX$ направлена на восток, а ось $ОY$ – на север. Пусть юго-западный угол плота имеет координаты $(x_1, y_1)$
, северо-восточный угол – координаты $(x_2, y_2)$.
Пловец находится в точке с координатами $(x, y)$. 

Определите, к какой стороне плота (северной, южной, западной или восточной) или к какому углу плота (северо-западному, северо-восточному, юго-западному, юго-восточному) пловцу нужно плыть, чтобы как можно скорее добраться до плота.

![text](images/1A-task.png)

**Формат ввода**

Программа получает на вход шесть чисел в следующем порядке: 
$x_1$, $y_1$ (координаты юго-западного угла плота), $x_2$, $y_2$ (координаты северо-восточного угла плота), $x$, $y$ (координаты пловца). Все числа целые и по модулю не превосходят 100. Гарантируется, что координаты пловца находятся вне плота.

**Формат вывода**

Если пловцу следует плыть к северной стороне плота, программа должна вывести символ ”N”, к южной — символ ”S”, к западной — символ ”W”, к восточной — символ ”E”. Если пловцу следует плыть к углу плота, нужно вывести одну из следующих строк: ”NW”, ”NE”, ”SW”, ”SE”.

````python
with open("input/1A_input.txt", 'r') as f:
    x1, y1, x2, y2, x, y = map(int, f.readlines())

# Последовательно рассмотрим возможные варианты расположения пловца 
# (условие задачи позволяет нам не обрабатывать некорректные варианты координат)

# Если пловец находится напротив плота по вертикали
if x1 < x < x2:
    res = 'N' if y >= y2 else 'S'
# Если пловец находится напротив плота по горизонтали
elif y1 < y < y2:
    res = 'W' if x <= x1 else 'E'
# Если пловец находится севернее плота
elif y >= y2:
    res = 'NW' if x <= x1 else 'NE'
# Если пловец находится южнее плота
else:
    res = 'SW' if x <= x1 else 'SE'

print(res)
````

### B. Майки и носки

Как известно, осенью и зимой светает поздно, и так хочется утром ещё хоть немного поспать, а не идти в школу! Некоторые школьники готовы даже одеваться, не открывая глаз, лишь бы отложить момент пробуждения. Вот и Саша решил, что майку и носки он вполне может вытащить из шкафа на ощупь с закрытыми глазами и только потом включить свет и одеться. 

В шкафу у Саши есть два ящика. В одном из них лежит $A$ синих и $B$ красных маек, в другом — $C$ синих и $D$ красных пар носков. Саша хочет, чтобы и майка, и носки были одного цвета. Он вслепую вытаскивает $M$ маек и $N$ пар носков. В первое же утро Саша задумался, какое минимальное суммарное количество предметов одежды ($M+N$) он должен вытащить, чтобы среди них гарантированно оказались майка и носки одного цвета. Какого именно цвета окажутся предметы одежды, для Саши совершенно неважно.

**Формат ввода**

На вход программе подаются четыре целых неотрицательных числа $A$ (количество синих маек), $B$ (количество красных маек), $C$ (количество синих пар носков) и $D$ (количество красных пар носков), записанных в отдельных строках. Гарантируется, что в шкафу есть одноцветный комплект из майки и носков.

**Формат вывода**

Программа должна вывести два числа: количество маек $M$ и количество пар носков $N$, которые должен взять Саша. Необходимо, чтобы обязательно нашлась одноцветная пара, при этом сумма $M+N$ должна быть минимальной.

````python
with open("input/1B_input.txt", 'r') as f:
    a, b, c, d = map(int, f.readlines())

# Если отсутствуют нули (то есть в ящиках лежат и майки, и носки обоих цветов)
if all([a, b, c, d]):
    max_tshirts = max(a, b)
    max_socks = max(c, d)
    min_same_color = min(a + c, b + d)

    if min(max_tshirts, max_socks) < min_same_color:
        if max_tshirts < max_socks:
            res = (max_tshirts + 1, 1)
        else:
            res = (1, max_socks + 1)
    else:
        res = (b + 1, d + 1) if b + d < a + c else (a + 1, c + 1)
# Если присутствуют нули
else:
    if a == 0 or c == 0:
        res = (a + 1, c + 1)
    elif b == 0 or d == 0:
        res = (b + 1, d + 1)

print(*res)
````

### C. Надпись на табло

**Решение**

Идея состоит в "сжатии" поступающих на вход матриц (символов):
1. Сразу же на этапе ввода данных игнорируем повторяющиеся подряд строки. И заодно игнориуем начальные и конечные пустые строки (под пустыми понимаются строки, в которых нет включеных диодов). 

2. Если во время чтения данных нам попадется пустая строка до и после которой будут непустые, мы сразу возвращаем `X` (поскольку у нас нет букв, состоящих из нескольких непересекающихся фигур).

3. Транспонируем матрицу, которая к этому моменту уже "сжата" по строкам и выполняем сжатие по столбцам.

4. В итоге полученное сжатое "изображение" сравниваем с сжатыми изображениями букв, которые нам необходимо распознать. Если не одна буква не подошла, возвращаем `X`

````python
def main():
    lines = []
    # Сразу же на этапе ввода данных игнорируем повторяющиеся подряд строки. 
    # И заодно игнориуем начальные и конечные пустые строки
    with open("input/1C_input.txt", 'r') as f:
        n = int(f.readline())
        state = 0
        while (line := f.readline()[: n]):
            # Если в строке нет включенных диодов
            if not '#' in line:
                # и при этом до этого были строки с включенными диодами,
                if state == 1:
                    # установим статус -1
                    state = -1
                    continue
                # Иначе просто пропускаем пустую (выключенную) строку
                else:
                    continue
            # Если в строке есть включенные диоды
            else:
                # и при этом значение статуса 
                # говорит о том, что ей предшествовала пустая 
                # строка-разделитель, то данный пример не соответствует 
                # ни одной из рассматриваемых букв (поскольку у нас нет букв,
                # состоящих из нескольких непересекающихся фигур)
                if state == -1:
                    print('X')
                    return
                
                if state == 0:
                    state = 1
                    lines.append(line + '.')
                elif line != lines[-1][: -1]:
                    lines.append(line + '.')
    # Если все строки были пустыми, дальше можно не продолжать
    if len(lines) == 0:
        print('X')
        return

    # Теперь транспонируем "сжатую" по строкам матрицу и выполним "сжатие" еще раз
    lines = list(map(list, lines))
    lines = list(zip(*lines))
    lines = [''.join(pair[0]) for pair in zip(lines, lines[1:]) if (pair[0] != pair[1]) and '#' in pair[0]]

    # Выполним сравнение получившегося "сжатого" изображения с соответствующими 
    # сжатыми изображениями интересующих нас букв (также в транспонированном виде, 
    # чтобы не выполнять еще одно преобразование)
    letters = {
        'P': ['####',
              '#.#.',
              '###.'], 
        'I': ['#'], 
        'O': ['###',
              '#.#',
              '###'], 
        'C': ['###',
              '#.#'], 
        'L': ['##',
              '.#'], 
        'H': ['###',
              '.#.',
              '###']
    }
    for l in letters:
        if len(lines) != len(letters[l]) or any([row != line for row, line in zip(letters[l], lines)]):
            continue
        else:
            print(l)
            return
    print('X')
      
main()
````

### D. Кондиционер (составление тестов)

На соревновании участникам была предложена следующая задача:

> В офисе, где работает программист Петр, установили кондиционер нового типа. Этот кондиционер отличается особой простотой в управлении. У кондиционера есть всего лишь два управляемых параметра: желаемая температура и режим работы.
>
> Кондиционер может работать в следующих четырех режимах:
>
> - «freeze» — охлаждение. В этом режиме кондиционер может только уменьшать температуру. Если температура в комнате и так не больше желаемой, то он выключается.
>
> - «heat» — нагрев. В этом режиме кондиционер может только увеличивать температуру. Если температура в комнате и так не меньше желаемой, то он выключается.
>
> - «auto» — автоматический режим. В этом режиме кондиционер может как увеличивать, так и уменьшать температуру в комнате до желаемой.
>
> - «fan» — вентиляция. В этом режиме кондиционер осуществляет только вентиляцию воздуха и не изменяет температуру в комнате.
>
> Кондиционер достаточно мощный, поэтому при настройке на правильный режим работы он за час доводит температуру в комнате до желаемой.
>
>Требуется написать программу, которая по заданной температуре в комнате troom, установленным на кондиционере желаемой температуре tcond и режиму работы определяет температуру, которая установится в комнате через час.

Вам предстоит разработать набор тестов (только входных данных) для этой задачи, тщательно проверяющий решения участников.

**Формат вывода**

Сдавать следует не программу, а текстовый файл.
В первой строке файла запишите число $N$ ($1 \leq N \leq 20$) — количество тестов, которые вы разработали.
В следующих $N$ строках запишите по одному тесту. Каждый тест должен состоять из чисел $t_{room}$ и $t_{cond}$ ($-50 \leq t_{room}, t_{cond} \leq 50$) и режима работы кондиционера (одно из слов *freeze*, *heat*, *auto*, *fan*).

**Пример формата файла** для сдачи:
```txt

10 20 heat
20 25 freeze
20 20 fan
```


**Решение**

Тесты должны охватить все 4 режима работы кондиционера и для каждого режима необходимо протестировать 3 ситуации:
- текущая температура равна желаемой температуре
- текущая температура больше желаемой
- текущая температура меньше желаемой

Таким образом, нам необходимо провести 12 тестов. Файл с тестами может быть таким:

